```python
Nsamp=12
theta=1.05*np.pi/180  # magic angle 

c6sym=True
umkl=0

l=MoireLattice.MoireTriangLattice(Nsamp,theta,0,c6sym,umkl) 
lq=MoireLattice.MoireTriangLattice(Nsamp,theta,2,c6sym,umkl) #this one 
[KX,KY]=lq.Generate_lattice()
Npoi=np.size(KX); print(Npoi, "numer of sampling lattice points")
[q1,q2,q3]=l.q
q=la.norm(q1)
umkl=1
print(f"taking {umkl} umklapps")
VV=lq.boundary()


a_graphene=2.458*(1e-10) #in meters this is the lattice constant NOT the carbon-carbon distance
A1mbz=lq.VolMBZ*((q**2)/(a_graphene**2))
ev_Ambz=1/A1mbz

df_prime=df.loc[df['nu']==0]
scatter(df_prime['kx'],df_prime['ky'], c=np.real(df_prime['bub']*ev_Ambz))
plt.plot(VV[:,0],VV[:,1], c='r')
colorbar()

plt.show()


df_one_temp=df.loc[(df['nu'] == 0.0) & (df['kx'] == 0.0)]
scatter(df_one_temp['ky'],df_one_temp['bub']*ev_Ambz)
scatter(df_one_temp['ky'],np.real(df_one_temp['bub']*ev_Ambz))
plt.show()

df_one_temp=df.loc[(df['nu'] == 0.0) & (df['ky'] == 0.0)]
plt.scatter(df_one_temp['kx'],df_one_temp['bub']*ev_Ambz)
```
